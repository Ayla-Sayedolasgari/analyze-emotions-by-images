```python
!pip install -q matplotlib transformers pillow google-colab

from transformers import pipeline
from PIL import Image
from google.colab import files
import matplotlib.pyplot as plt

# -----------------------------
# ÿ®ÿßÿ±⁄Øÿ∞ÿßÿ±€å ŸÖÿØŸÑ
# -----------------------------
classifier = pipeline("image-classification", model="dima806/facial_emotions_image_detection")

# -----------------------------
# ÿ¢ŸæŸÑŸàÿØ ÿπ⁄©ÿ≥
# -----------------------------
print("üìå ŸÑÿ∑ŸÅÿßŸã €å⁄© ÿπ⁄©ÿ≥ ÿßŸÜÿ™ÿÆÿßÿ® Ÿà ÿ¢ŸæŸÑŸàÿØ ⁄©ŸÜ€åÿØ:")
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# ÿ®ÿßÿ≤ ⁄©ÿ±ÿØŸÜ Ÿà ŸÜŸÖÿß€åÿ¥ ÿπ⁄©ÿ≥
image = Image.open(image_path)
display(image)

# -----------------------------
# ÿ™ÿ≠ŸÑ€åŸÑ ÿßÿ≠ÿ≥ÿßÿ≥ÿßÿ™
# -----------------------------
results = classifier(image)

# ⁄ÜÿßŸæ ŸÜÿ™ÿß€åÿ¨ ÿØÿ± ⁄©ŸÜÿ≥ŸàŸÑ
print("\nüìå ŸÜÿ™ÿß€åÿ¨ ÿ™ÿ≠ŸÑ€åŸÑ ÿßÿ≠ÿ≥ÿßÿ≥ÿßÿ™:")
labels = []
scores = []

for res in results:
    print(f"{res['label']}: {res['score']:.4f}")
    labels.append(res['label'])
    scores.append(res['score'])

# -----------------------------
# ÿ±ÿ≥ŸÖ ŸÜŸÖŸàÿØÿßÿ± ÿØÿß€åÿ±Ÿá‚Äåÿß€å
# -----------------------------
plt.figure(figsize=(6,6))
plt.pie(scores, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title("results")
plt.show()
```
